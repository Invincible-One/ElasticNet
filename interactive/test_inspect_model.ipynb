```python
import sys
sys.path.append("..")

import torch
import torchvision
import torch.nn
from torchvision.models import resnet50


from models.custom_models import FullyConnectedNetwork, VGG_like



net1_input_d = 200 * 200
net2_input_d = (1, 200, 200)
net1_config = [2048,]
net2_config = [32, 'M', 64, 'M', 128, 'M', 256, 'M', 512, 'M', 1024, 'M', 1024, 'M', 1024]


network1 = FullyConnectedNetwork(net1_input_d, net1_config)
#network2 = VGG_like(net2_input_d, net2_config)
#network3 = resnet50(weights=None)
```

```python
# FullyConnectedNetwork
print("Names and Modules:")
for name, module in network1.named_children():
    print(name, module)

print("\nOnly Modules:")
for module in network1.children():
    print(module)

print("\nAll nested modules:")
for name, module in network1.named_modules():
    print(name, module)
```
