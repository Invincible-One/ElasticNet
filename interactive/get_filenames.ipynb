#### Get filenames for job scripts of finetune

```python
import os
import glob


def remove_warning_first_line(folder, pattern, warning_message="WARNING: Could not find any nv files on this host!"):
    for fpath in glob.glob(os.path.join(folder, pattern)):
        with open(fpath, 'r') as f:
            lines = f.readlines()
        if lines and lines[0].strip() == warning_message:
            with open(fpath, 'w') as f:
                f.writelines(lines[1:])


def extract_filenames(folder, pattern):
    substrings = []
    for fpath in glob.glob(os.path.join(folder, pattern)):
        try:
            with open(fpath, 'r') as f:
                for i, line in enumerate(f):
                    if i == 2:
                        substrings.append('"' + line[17 : -1] + '"')
                        break
        except Exception as e:
            print(f"Error reading {fpath}: {e}!")
    
    print(f"n_files: {len(substrings)}")
    return '(' + ' '.join(substrings) + ')'
```

```python
folder = "/home/ym2380/elastic_net/official/logs/exp11/shallow_net/config1024/reg"
pattern = "e_*.out"

remove_warning_first_line(folder, pattern)
res = extract_filenames(folder, pattern)

print(res)
```
